```python
from scipy import interpolate
import numpy as np
import matplotlib.pyplot as plt
```

```python
points = [[0,0], 
          [0,0.2],
          [0.5,0.1],
          [-0.1,1]]

points.append(points[0]) # make it a closed loop
points = np.array(points)
xp = points[:,0]
yp = points[:,1]
t = np.linspace(0, 1, len(xp))
xf = interpolate.interp1d(t, xp, assume_sorted=True)
yf = interpolate.interp1d(t, yp, assume_sorted=True)
```

```python
for t in np.linspace(0, 1, 1000):
    plt.plot(xf(t), yf(t), ".k")
```
