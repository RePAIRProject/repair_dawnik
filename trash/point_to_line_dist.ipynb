```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def dist_point_to_line(p, a, b):
    p = np.array(p)
    a = np.array(a)
    b = np.array(b)
    ab = b - a
    D  = ab / np.linalg.norm(ab)
    return np.abs(np.sum(D*(p - a)))

def dist_point_to_line2(p, a, b):
    p = np.array(p)
    a = np.array(a)
    b = np.array(b)

    ba = b - a
    pa = a - p
    c = np.sum(ba*pa)
    if c > 0:
        return np.sqrt(np.sum(pa*pa))
    bp = p - b
    if np.sum(ba*bp) > 0:
        return np.sqrt(np.sum(bp*bp))
    e = pa - ba * (c / np.sum(ba*ba))
    return np.sqrt(np.sum(e*e))
```

```python
a = [1,1]
b = [1,2]

N = 200
M = 200
mat = np.zeros((N,M))
xx = np.linspace(0, 3, N)
yy = np.linspace(0, 3, M)
for i,x in enumerate(xx):
    for j,y in enumerate(yy):
        p = [x,y]
        mat[i,j] = dist_point_to_line2(p, a, b)

xx,yy = np.meshgrid(xx,yy, indexing="ij")
plt.pcolormesh(xx,yy,mat, vmin=0, vmax=2)
plt.plot(a[0], a[1], "o")
plt.plot(b[0], b[1], "o")
```
